## K Nearet Neighbour [KNN] Classification

The K-Nearest Neighbors (KNN) algorithm is a supervised machine learning method employed to tackle classification and regression problems.

K-NN is a non-parametric algorithm that assumes the similarity between the new case/data and available cases and put the new case into the category that is most similar to the available categories.

It is also called a lazy learner algorithm because it does not learn from the training set immediately instead it stores the dataset and at the time of classification, it performs an action on the dataset.

##### **Example**

Loading required modules

```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
```

Loading Dataset

```python
iris = datasets.load_iris()

print(iris.DESCR)
```

Features & Labels

```python
features = iris.data
labels = iris.target

print(features[0])
print(labels[0])
```

Training the classifier

```python
clf = KNeighborsClassifier()

clf.fit(features, labels)
```

Predictions

```python
pred = clf.predict([[1, 1, 1, 1]])

predicted_species = iris.target_names[pred[0]]           # Map the numeric prediction to the species name

print(pred)
print(predicted_species)
```

Determining that boundary is called Classfication.

**Example of how a KNN Classifier works**

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter([1, 2, 3, 4, 6, 2, 1.5], [3, 4, 6, 5, 7, 5, 6], label='Category A')
ax.scatter([3, 5, 5, 6, 6, 4.5, 5], [5, 4, 6, 5, 6, 7, 5], label='Category B')

ax.scatter([4.2], [6])


area1 = plt.Circle((4.2, 6), 0.5, fill=False)
area2 = plt.Circle((4.2, 6), 1, fill=False)

# add the circle to the plot
ax.add_patch(area1)
ax.add_patch(area2)

# ensure the circles are fully visible on the plot.
ax.set_xlim(0, 7)
ax.set_ylim(2.5, 8)

plt.legend()
plt.show()
```

To determine whether the green dot belongs to category A or B, a boundary/area is drawn around it with some value of _k_(radius).

In that area, if 

    Category A points > Category B points   --->   Green is A

and if
  
    Category A points < Category B points   --->   Green is B   

Here, we have 2 B points and 1 A, so Green will said to be B.
